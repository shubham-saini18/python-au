```python
def greeting(): 
    x="Good Morning"
    return x
```

```python
def custom_name(name):
    greet= greeting()+f" {name} !!"
    return greet
    
```

```python
custom_name("satya")
```

```python
def square(x):
    return x*x
list45=[2,3,4,5,6,7,8]   
```

```python
list(map(square,list45))
```

```python
list(map(lambda x:"even" if x%2==0 else "odd",list45))
```
