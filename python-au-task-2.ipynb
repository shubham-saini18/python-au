## Problem 1: Write a program that will give you in hand monthly salary
    after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes
    deduction as below:
    Salary(Lakhs) : Tax(%)
    • Below 5 : 0%
    • 5-10 : 10%
    • 10-20 : 20%
    • aboove 20 : 30%

```python
# Program to calculate in-hand monthly salary from CTC
ctc = float(input("Enter your CTC in Lakhs: "))

# Convert lakhs to rupees
annual_salary = ctc * 100000

# Deductions
hra = 0.10 * annual_salary
da = 0.05 * annual_salary
pf = 0.03 * annual_salary

# Tax calculation based on slabs
if ctc < 5:
    tax_percent = 0
elif ctc <= 10:
    tax_percent = 10
elif ctc <= 20:
    tax_percent = 20
else:
    tax_percent = 30

tax_amount = (tax_percent / 100) * annual_salary

# In-hand annual salary
inhand_annual = annual_salary - (hra + da + pf + tax_amount)

# Monthly salary
monthly_salary = inhand_annual / 12

print("----- Salary Breakdown -----")
print(f"Annual Salary: ₹{annual_salary}")
print(f"HRA Deduction (10%): ₹{hra}")
print(f"DA Deduction (5%): ₹{da}")
print(f"PF Deduction (3%): ₹{pf}")
print(f"Tax ({tax_percent}%): ₹{tax_amount}")
print(f"In-hand Monthly Salary: ₹{monthly_salary}")
```

## Problem 2: Write a program that take a user input of three angles and will find out whether it can form a triangle or not.


```python
# Program to check if three angles form a triangle

a = float(input("Enter angle 1: "))
b = float(input("Enter angle 2: "))
c = float(input("Enter angle 3: "))

# Check triangle condition
if a > 0 and b > 0 and c > 0 and (a + b + c == 180):
    print("The angles form a valid triangle.")
else:
    print("The angles do NOT form a triangle.")
```

## Problem 3: Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit.

```python
# Program to check Profit or Loss

cp = float(input("Enter Cost Price: "))
sp = float(input("Enter Selling Price: "))

if sp > cp:
    print(f"Profit of ₹{sp - cp}")
elif sp < cp:
    print(f"Loss of ₹{cp - sp}")
else:
    print("No Profit, No Loss.")
```

## Problem 4: Write a menu-driven program -
    1. cm to ft
    2. km to miles
    3. USD to INR
    4. exit


```python
while True:
    print("----- MENU -----")
    print("1. cm to ft")
    print("2. km to miles")
    print("3. USD to INR")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        cm = float(input("Enter value in cm: "))
        ft = cm / 30
        print(f"{cm} cm = {ft} ft")

    elif choice == 2:
        km = float(input("Enter value in km: "))
        miles = km * 0.6
        print(f"{km} km = {miles} miles")

    elif choice == 3:
        usd = float(input("Enter amount in USD: "))
        inr = usd * 88.66  
        print(f"${usd} = ₹{inr}")

    elif choice == 4:
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Please try again.")
```

## Problem 7 - Reverse a given integer number.

```python
# Program to reverse a given integer number

n = int(input("Enter an integer: "))
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print("Reversed number:", rev)
```

## Problem 8: Take a user input as integer N. Find out the sum from 1 to N.
    If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

#### Without loop

```python
# Program to calculate sum from 1 to N
# Skip numbers divisible by 5
# Stop if sum exceeds 300

N = int(input("Enter N: "))

i = 1
total = 0

while i <= N:
    if i % 5 != 0:       # skip numbers divisible by 5
        total += i

    if total > 300:      # stop further calculation
        break

    i += 1

print("Final Sum =", total)
```

#### With Loop

```python
N = int(input("Enter N: "))

total = 0

for i in range(1, N + 1):

    if i % 5 == 0:   # skip numbers divisible by 5
        continue

    total += i       # add to sum

    if total > 300:  # stop calculation immediately
        break

print("Final Sum =", total)
```

## Problem 9: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

```python
# Program to accept numbers until user enters 0
# Then display sum and average

total = 0
count = 0

while True:
    num = float(input("Enter a number (0 to stop): "))
    
    if num == 0:
        break
    
    total += num
    count += 1

if count == 0:
    print("No numbers were entered.")
else:
    average = total / count
    print("Sum =", total)
    print("Average =", average)
```

## Problem 9: Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
result = []

for n in range(2000, 3201):   # 3201 because upper bound is included
    if n % 7 == 0 and n % 5 != 0:
        result.append(str(n))

print(",".join(result))
```

## Problem 10: Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.

```python
result = []

for num in range(1000, 3001):
    num_str = str(num)
    
    if all(int(digit) % 2 == 0 for digit in num_str):
        result.append(num_str)

print(" ".join(result))
```

## Problem 11: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps. 
The trace of robot movement is
  shown as the following:
  UP 5
  DOWN 3
  LEFT 3
  RIGHT 2
  !
  The numbers after the direction are steps.
  ! means robot stop there.
  Please write a program to compute the distance from current position after a sequence of movement and original point.

```python
import math

x = 0
y = 0

while True:
    command = input()

    if command == "!":
        break

    direction, steps = command.split()
    steps = int(steps)

    if direction == "UP":
        y += steps
    elif direction == "DOWN":
        y -= steps
    elif direction == "LEFT":
        x -= steps
    elif direction == "RIGHT":
        x += steps

distance = math.sqrt(x*x + y*y)
print(round(distance))
```

## Problem 12:Write a program to print whether a given number is a prime number or not

```python
num = int(input("Enter the number for checking it prime or not: "))

if num < 2:
    print(f"{num} is not a prime number")
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            print(f"Entered Number {num} is Not a Prime Number")
            break
    else:
        print(f"Entered Number {num} is a Prime Number")
```

## Problem 13: Print all the Armstrong numbers in a given range. Range will be provided by the user Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

```python
start = int(input("Enter the starting number of range: "))
end = int(input("Enter the ending number of range: "))

print(f"Armstrong numbers between {start} and {end} are:")

for num in range(start, end + 1):
    sum_of_cubes = 0
    temp = num

    while temp > 0:
        digit = temp % 10
        sum_of_cubes += digit ** 3
        temp //= 10

    if sum_of_cubes == num:
        print(num)
```

## Problem 14: Calculate the angle between the hour hand and minute hand.
   Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final result angle. 
   For example, if the final angle is 10.61, we need to print 10.
   Input: H = 9 , M = 0 
   Output: 90 Explanation: The minimum angle between hour and minute hand when the time is 9 is 90 degress

```python
H = int(input("Enter hour (0-12): "))
M = int(input("Enter minute (0-59): "))

minute_angle = M * 6
hour_angle = 30 * H + 0.5 * M

angle = abs(hour_angle - minute_angle)
min_angle = min(angle, 360 - angle)

print(f"{int(min_angle)} degree")
```

## Problem 15: Given two rectangles, find if the given two rectangles overlap or not. A rectangle is denoted by providing the x and y coordinates of two points: the left top corner and the right bottom corner of the rectangle. Two rectangles sharing a side are considered overlapping. (L1 and R1 are the extreme points of the first rectangle and L2 and R2 are the extreme points of the second rectangle).

```python
def is_overlapping(L1, R1, L2, R2):

    # If one is completely to the left of the other
    if R1[0] < L2[0] or R2[0] < L1[0]:
        return False

    # If one is completely above the other
    if L1[1] < R2[1] or L2[1] < R1[1]:
        return False

    return True


# Input
L1 = tuple(map(int, input("Enter L1 (x1 y1): ").split()))
R1 = tuple(map(int, input("Enter R1 (x2 y2): ").split()))
L2 = tuple(map(int, input("Enter L2 (x3 y3): ").split()))
R2 = tuple(map(int, input("Enter R2 (x4 y4): ").split()))

# Output
if is_overlapping(L1, R1, L2, R2):
    print("Rectangles Overlap")
else:
    print("Rectangles Do Not Overlap")
```
