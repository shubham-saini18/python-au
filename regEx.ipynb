```python
import re
```

```python
with open("data.txt") as file:
    data = file.read()

names = re.findall(r"^[A-Z][a-zA-Z]+(?:\s[A-Z][a-zA-Z-]+)+", data, re.MULTILINE)
print("Names : \n")
for name in names:
    print(name)
```

```python
phones = re.findall(r"\d{3}-\d{3}-\d{4}", data)
for phone in phones:
    print(phone)
```

```python
emails = re.findall(r"[\w\.-]+@[\w\.-]+\.\w+", data)
for email in emails: 
    print(email)
```

```python
addresses = re.findall(r"\d+\s[\w\s\.]+,\s[\w\s]+\s[A-Z]{2}\s\d{5}", data)
for address in addresses:
    print(address)
```

```python
names_A = re.findall(r"^A[a-zA-Z]+(?:\s[A-Z][a-zA-Z-]+)+", data, re.MULTILINE)

print("Names starting with A:\n")
for name in names_A:
    print(name)
```

```python
names_B = re.findall(r"^B[a-zA-Z]+(?:\s[A-Z][a-zA-Z-]+)+", data, re.MULTILINE)

print("Names starting with A:\n")
for name in names_B:
    print(name)
```
