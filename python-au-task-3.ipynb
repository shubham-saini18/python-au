## Problem 5: Write a Python Program to Find the Sum of the Series till the nth term:
    1 + x^2/2 + x^3/3 + â€¦ x^n/n ,n will be provided by the user

```python
X = float(input("Enter the value of X: "))
n = int(input("Enter the value of n: "))
sum_series = 1
for i in range(2, n + 1):
    term = (X ** i) / i
    sum_series += term
print(f"Sum of the series = {sum_series}")
```

## Problem 6: The natural logarithm can be approximated by the following series.
## (x-1)/x + (1/2)((x-1)/x)^2 + (1/2)((x-1)/x)^3 + (1/2)((x-1)/x)^4.........
    If x is input through the keyboard, write a program to calculate the sum of the first seven terms of this series.

```python
x1 = float(input("Enter the value of x: "))
n1 = int(input("Enter number of terms (n): "))

u = (x1 - 1) / x1 
sum_series = u      
for i in range(2, n1 + 1):
    term = 0.5 * (u ** i)
    sum_series += term

print(f"Sum of {n1} terms = {sum_series}")
```

## Problem 9: Write a program that will take a decimal number as input and prints out the binary equivalent of the number

### Method-1 (Simple method by using remainder)

```python
num = int(input("Enter a decimal number: "))

binary = ""
n = num

if n == 0:
    binary = "0"
else:
    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n //= 2

print("Binary equivalent of", num, "is:", binary)
```

### Method-2 (By using bin() )

```python
num = int(input("Enter a decimal number: "))
print("Binary:", bin(num)[2:])
```

### Method-3 (collect remainder in list, then reverse the list)

```python
num = int(input("Enter a decimal number: "))
n = num
bits = []
if n == 0:
    print("Binary:", 0)
else:
    while n > 0:
        bits.append(n % 2)   
        n //= 2              
    bits.reverse()           
    for bit in bits:
        print(bit, end="")
```

## Problem 10: Write a program that will take 2 numbers as input and prints the LCM and HCF of those 2 numbers


### Method 1: By using Loop

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# HCF
hcf = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        hcf = i

# LCM
lcm = (a * b) // hcf

print("HCF =", hcf)
print("LCM =", lcm)
```

### Method 2: Subtraction Method

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

x, y = a, b

while x != y:   
    if x > y:          
        x = x - y      
    else:
        y = y - x   

hcf = x
lcm = (a * b) // hcf  

print("HCF =", hcf)
print("LCM =", lcm)
```

## Problem 11: Create Short Form from initial character.
    Given a string create short form ofthe string from Initial character. Short form should be capitalised.
    Example:
       Input: Data science mentorship program
       Output: DSMP


#### METHOD 1: Character-by-character logic

```python
text = "Data science mentorship program "
short = ""

for i in range(len(text)):
    if i == 0 or text[i-1] == " ":
        short += text[i].upper()

print(short)
```

### METHOD 2: Using a flag

```python
text = "Data science mentorship program"
short = ""
flag = True

for ch in text:
    if flag:
        short += ch.upper()
    flag = (ch == " ")

print(short)
```

## Problem 12: Append second string in the middle of first string
     Input: campusx
            data
     Output: camdatapusx

### METHOD - 1: By using Slicing

```python
str1 = "campusx"
str2 = "data"

mid_point = len(str1) // 2
result = str1[:mid_point] + str2 + str1[mid_point:]

print(result)
```

### METHOD - 2: By using for-Loop  

```python
str1 = "campusx"
str2 = "data"

mid_point = len(str1) // 2
result = ""

for i in range(len(str1)):
    if i == mid_point:
        result += str2
    result += str1[i]

print(result)
```

## Problem 13:Given string contains a combination of the lower and upper case letters. Write a program to arrange the characters of a string so that all lowercase letters should come first.
    Given: str1 = PyNaTive
    Expected Output: yaivePNT

### Method-1: for-loop

```python
str = "PyNaTive"
lower = ""
upper = ""
for char in str:
    if char.islower():
        lower += char
    else:
        upper += char
result = lower + upper
print(result)
```

### Method-2: Using list comprehension

```python
str = "PyNaTive"
result = "".join([char for char in str if char.islower()]) + "".join([char for char in str if char.isupper()])
print(result)
```

## Problem 14: Take a alphanumeric string input and print the sum and average of the digits that appear in the string, ignoring all other characters.
    Input: hel123O4every093
    Output: Sum: 22
            Avg: 2.75

```python
str = "hel123O4every093"

total = 0
count = 0

for char in str:
    if char.isdigit():
        total += int(char)
        count += 1
      
avg = total / count

print("Sum:", total)
print("Avg:", avg)
```
