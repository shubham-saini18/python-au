## Problem 5: Write a Python Program to Find the Sum of the Series till the nth term:

## Problem 9: Write a program that will take a decimal number as input and prints out the binary equivalent of the number

### Method-1 (Simple method by using remainder)

```python
num = int(input("Enter a decimal number: "))

binary = ""
n = num

if n == 0:
    binary = "0"
else:
    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n //= 2

print("Binary equivalent of", num, "is:", binary)
```

### Method-2 (By using bin() )

```python
num = int(input("Enter a decimal number: "))
print("Binary:", bin(num)[2:])
```

### Method-3 (collect remainder in list, then reverse the list)

```python
num = int(input("Enter a decimal number: "))
n = num
bits = []
if n == 0:
    print("Binary:", 0)
else:
    while n > 0:
        bits.append(n % 2)   
        n //= 2              
    bits.reverse()           
    for bit in bits:
        print(bit, end="")
```

## Problem 10: Write a program that will take 2 numbers as input and prints the LCM and HCF of those 2 numbers


### Method 1: By using Loop

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# HCF
hcf = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        hcf = i

# LCM
lcm = (a * b) // hcf

print("HCF =", hcf)
print("LCM =", lcm)
```

### Method 2: Subtraction Method

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

x, y = a, b

while x != y:   
    if x > y:          
        x = x - y      
    else:
        y = y - x   

hcf = x
lcm = (a * b) // hcf  

print("HCF =", hcf)
print("LCM =", lcm)
```

## Problem 12: Append second string in the middle of first string
     Input: campusx
            data
     Output: camdatapusx

### METHOD - 1: By using Slicing

```python
str1 = "campusx"
str2 = "data"

mid_point = len(str1) // 2
result = str1[:mid_point] + str2 + str1[mid_point:]

print(result)
```

```python
str1 = "campusx"
str2 = "data"

mid_point = len(str1) // 2
result = ""

for i in range(len(str1)):
    if i == mid_point:
        result += str2
    result += str1[i]

print(result)
```

## Problem 13:Given string contains a combination of the lower and upper case letters. Write a program to arrange the characters of a string so that all lowercase letters should come first.
    Given: str1 = PyNaTive
    Expected Output: yaivePNT

### Method-1: for-loop

```python
str = "PyNaTive"
lower = ""
upper = ""
for char in str:
    if char.islower():
        lower += char
    else:
        upper += char
result = lower + upper
print(result)
```

### Method-2: Using list comprehension

```python
str = "PyNaTive"
result = "".join([char for char in str if char.islower()]) + "".join([char for char in str if char.isupper()])
print(result)
```

## Problem 14: Take a alphanumeric string input and print the sum and average of the digits that appear in the string, ignoring all other characters.
    Input: hel12304every093
    Output: Sum: 22
            Avg: 2.75

```python
str = "hel12304every093"

total = 0
count = 0

for char in str:
    if char.isdigit():
        total += int(char)
        count += 1
      
avg = total / count

print("Sum:", total)
print("Avg:", avg)
```

## Problem 16: Check whether the string is Symmetrical.
### Statement: Given a string. the task is to check if the string is symmetrical or not. A string is said to be symmetrical if both the halves of the string are the same.
    Example 1:
     Input:- khokho
     Output:- The entered string is symmetrical

#### Method-1: By using Loop

```python
s = "khokho"
n = len(s)
if n % 2 != 0:
    print("The entered string is not symmetrical")
else:
    mid = n // 2
    flag = True
    for i in range(mid):
        if s[i] != s[mid + i]:
            flag = False
            break
    if flag:
        print("The entered string is symmetrical")
    else:
        print("The entered string is not symmetrical")
```

#### Method-2 : By using slicing

```python
s = "khokho"
mid = len(s) // 2              
first_half = s[:mid]            
second_half = s[mid:]
if first_half == second_half:
    print("The entered string is symmetrical")
else:
    print("The entered string is not symmetrical")
```

## Problem 17: Reverse words in a given String
### Statement: We are given a string and we need to reverse words of a given string.
    Example 1:
      Input: geeks quiz practice code
      Output: code practice quiz geeks
    Example 2:
      Input: my name is laxmi
      Output: laxmi is name my

#### Method-1 : Using split() and join()

```python
s = "geeks quiz practice code"
words_split = s.split()
rev_words_split = words_split[::-1]
rev_s = " ".join(rev_words_split)
print(rev_s)
```

#### Method-2 : Using a for-loop + reversed()

```python
s = "geeks quiz practice code"
words = s.split()
rev_s = ""
for word in reversed(words):
    rev_s += word + " "

print(rev_s)
```

#### Method-3 : Using a while-loop + indexes

```python
str = "geeks quiz practice code"
words = s.split()
rev_str = ""
i = len(words) - 1
while i >= 0:
    rev_str += words[i] + " "
    i -= 1
print(rev_str)
```

## Problem 18: Find uncommon words from two Strings.
### Statement: Given two sentences as strings A and B. The task is to return a list of all uncommon words. A word is uncommon if it appears exactly once in any one of the sentences, and does not appear in the other sentence. Note: A sentence is a string of space-separated words. Each word consists only of lowercase letters.
    Example 1:
      Input: A = "apple banana mango"
             B = "banana fruits mango"
      Output: ['apple', 'fruits']

#### Method-1 : Compare lists directly

```python
A = "apple banana mango"
B = "banana fruits mango"
wordsA = A.split()
wordsB = B.split()
uncommon = []
for word in wordsA:
    if word not in wordsB:
        uncommon.append(word)

for word in wordsB:
    if word not in wordsA:
        uncommon.append(word)
print(uncommon)
```

#### Method-2 : Using Counter

```python
from collections import Counter
A = "apple banana mango"
B = "banana fruits mango"
count = Counter(A.split()) + Counter(B.split())
uncommon = [word for word, freq in count.items() if freq == 1]
print(uncommon)
```

#### Method-3 : Using Loop

```python
A = "apple banana mango"
B = "banana fruits mango"
split_A = A.split()
split_B = B.split()
uncommon = []

for w in split_A:
    if split_A.count(w) == 1 and w not in split_B:
        uncommon.append(w)
        
for w in split_B:
    if split_B.count(w) == 1 and w not in split_A:
        uncommon.append(w)
print(uncommon)
```

## Problem 19: Word location in String.
### Statement: Find a location of a word in a given sentence.
    Example 1:
      Input:
        Sentence: We can learn data science through campusx mentorship program.
        word: campusx
      Output: Location of the word is 7.
## Note- Don't use index/find functions 

### Method 1: Using Loop + Counter

```python
sentence = "We can learn data science through campusx mentorship program"
word = "campusx"

words = sentence.split()

position = 1
found = False
for w in words:
    if w == word:
        print("Location of the word is", position)
        found = True
        break
    position += 1

if not found:
    print("Word not found")
```

### Method 2: Using while-loop

```python
sentence = "We can learn data science through campusx mentorship program"
word = "campusx"

words = sentence.split()
i = 0

while i < len(words):
    if words[i] == word:
        print("Location of the word is", i+1)
        break
    i += 1
```
